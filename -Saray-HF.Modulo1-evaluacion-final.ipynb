2.7 Realizar_compra(): Permite a un cliente realizar una compra seleccionando productos del
inventario.

```python
def realizar_compra():
    carrito = []
    coste_total = 0

    while True:
        print("Inventario disponible:")
        for producto in inventario_tienda:
            print(f"Item: {producto['item']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")

        compra = input("¿Qué desea añadir a su carrito? (Escriba 'nada' para terminar):")

        if compra.lower() == "nada":
            break

        
        for producto in inventario_tienda:
            if producto['item'].lower() == compra.lower():
                carrito.append(producto['item'])
                coste_total += producto['precio']
                print(f"{producto['item']} añadido al carrito.")
                break
        else:
            
            print("Producto no encontrado en el inventario.")

    print("Carrito final:", carrito)
    print("Coste total:", coste_total)
```

```python
realizar_compra ()
```
